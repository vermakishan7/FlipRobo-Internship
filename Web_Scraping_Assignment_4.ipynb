```python
#Creating a session
driver = webdriver.Chrome()
driver.get('https://www.bcci.tv/')
time.sleep(4)

#getting the link for international fixture page and going on that page
a=driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down'][1]").get_attribute('href')
driver.get(a)
time.sleep(4)

#Creating empty lists
date=[]
title=[]
series=[]
place=[]

#scraping match titles
for i in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):
    title.append(i.text)
    
#scraping date
for i in driver.find_elements_by_xpath("//div[@class='fixture__full-date']"):
    date.append(i.text)
    
#scraping series name
for i in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']"):
    series.append(i.text)
    
#scraping place of match
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    place.append(i.text)
    
#creating dataframe and storing the scraped datain it
df=pd.DataFrame({'Match Title':title,'Series':series,'Place':place,'dt':date,'Date':date,'Time':date})

#spliting date and time
df['dt']=df['dt'].str.replace('\n',',')
df['dt']=df['dt'].str.split(',')
for i in range(0,len(df['dt'])):
    df['Date'][i]=" ".join(df['dt'][i][:2])
for i in range(0,len(df['dt'])):
    df['Time'][i]=df['dt'][i][-1]
    
#droping unwanted columns
df.drop(columns='dt',inplace=True)

#Showing the result
df
```

```python
#creating a session
driver = webdriver.Chrome()
driver.get('https://github.com/')

#clicking on explore menu
driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]").click()

#clicking on trending option from explore menu
driver.find_element_by_xpath("//div[@class='dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul[2]/li[3]").click()

#creating empty lists to store data
title=[]
desc=[]
count=[]
lang=[]
urls=[]

#getting urls of the trending repositories
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    urls.append(i.get_attribute('href'))

#Going to each page and scraping the required data
for i in urls:
    driver.get(i)
    time.sleep(4)
    
    #scraping title using xpath
    try:
        title.append(driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a").text)
    except NoSuchElementException :
        title.append("-")
    
    #scraping description using xpath
    try:
        desc.append(driver.find_element_by_xpath("//p[@class='f4 mt-3']").text)
    except NoSuchElementException :
        desc.append("-")
    
    #scraping contributors using xpath
    try:
        count.append(driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/h2/a/span").text)
    except NoSuchElementException :
        count.append("-")
    
    #scraping language used 
    try:
        lang.append(driver.find_element_by_xpath("//ul[@class='list-style-none']/li[1]/a/span[1]").text)
    except NoSuchElementException :
        lang.append("-")
        
#creating dataframe and storing data in it
df=pd.DataFrame({'Repository title':title,'Repository description':desc,'Contributors count':count,'Language used':lang})

#results
df
```

```python
#creating a session
driver = webdriver.Chrome()
driver.get('https://www.billboard.com')
time.sleep(4)

#clicking on charts option
driver.find_element_by_xpath("//header[@class='header header--desktop padding-x--sm flex--space-between-center bg--dark']/div/ul[1]/li[1]").click()
time.sleep(4)

#clicking on hot 100 page
driver.find_element_by_xpath("//div[@class='all-charts-list']/div[2]/a[1]").click()
time.sleep(4)

#Creating empty lists
s_name=[]
a_name=[]
lw_rank=[]
peek_rank=[]
weeks_onboard=[]

#scraping song name using xpath
for i  in driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']"):
    s_name.append(i.text)
    
#scraping artist name using xpath
for i  in driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']"):
    a_name.append(i.text)

#scraping last week rank using xpath    
for i  in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']"):
    lw_rank.append(i.text)
    
#scraping peek rank using xpath
for i  in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']"):
    peek_rank.append(i.text)
    
#scraping weeks onboard using xpath
for i  in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']"):
    weeks_onboard.append(i.text)
    
#creating dataframe
df=pd.DataFrame({"Song name":s_name,'Artist name':a_name,'Last week rank':lw_rank,'Peek rank':peek_rank,'Weeks onboard':weeks_onboard})

#output
df
```

```python
driver = webdriver.Chrome()
driver.get('https://www.naukri.com/')
time.sleep(4)

#clicking on recruiters option
driver.find_element_by_xpath("//a[@title='Search Recruiters']").click()
time.sleep(2)

#switching focus to the new window i.e. recruiters window
p = driver.current_window_handle
chwd = driver.window_handles
for w in chwd:
    if(w!=p):
        driver.switch_to.window(w)
        break
time.sleep(2)

#entering Data Sceience in recruiters search bar
driver.find_element_by_xpath("//input[@class='sugInp']").send_keys('Data Science')

#clicking on search button
driver.find_element_by_id("qsbFormBtn").click()

#fetching urls of each recruiter
urls=[]
for i in driver.find_elements_by_xpath("//div[@class='recImg mid_pImg_Silht']/a"):
    urls.append(i.get_attribute('href'))
    
#creating empty lists
name=[]
desig=[]
company=[]
skills=[]
loc=[]

#Going to each url
for i in urls:
    driver.get(i)
    
    #scraping name 
    try:
        name.append(driver.find_element_by_xpath("//h1[@class='fl ellipsis wLimit hd']").text)
    except NoSuchElementException :
        name.append("-")
        
    
    #scraping designation
    try:
        desig.append(driver.find_element_by_xpath("//div[@class='ellipsis']").text)
    except NoSuchElementException :
        desig.append("-")
    
    #scarping company name
    try:
        company.append(driver.find_element_by_xpath("//div[@class='oh'][1]/a").text)
    except NoSuchElementException :
        company.append("-")
     
    #scraping location
    try:
        loc.append(driver.find_element_by_xpath("//div[@class='oh'][2]/a").text)
    except NoSuchElementException :
        loc.append("-")
    
    #scraping skills hire for
    try:
        skills.append(driver.find_element_by_xpath("//div[@class='fl lPortn']").text)
    except NoSuchElementException :
        skills.append("-")
        
#creating dataframe
df=pd.DataFrame({'Name':name,'Designation':desig,'Company':company,'Location':loc,'Skills hire for':skills})

#output
df
```

```python
#creating a session
driver=webdriver.Chrome()
driver.get('https://www.imdb.com/list/ls095964455/')

#creating empty lists
name=[]
year_span=[]
genre=[]
runtime=[]
ratings=[]
votes=[]

#scraping name
for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
    name.append(i.text)

#scraping year span
for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]"):
    year_span.append(i.text)

    #scraping genre
for i in driver.find_elements_by_xpath("//span[@class='genre']"):
    genre.append(i.text)

#scraping runtime
for i in driver.find_elements_by_xpath("//span[@class='runtime']"):
    runtime.append(i.text)
    
#scraping ratings
for i in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]"):
    ratings.append(i.text)
    
#scraping votes
for i in driver.find_elements_by_xpath("//span[@name='nv']"):
    votes.append(i.text)

#creating dataframe
df=pd.DataFrame({'Name':name,'Year span':year_span,'Genre':genre,'Run time':runtime,'Ratings':ratings,'Votes':votes})

#output
df
```
