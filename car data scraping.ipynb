```python
# Fetching urls
urls = []

for i in driver.find_elements_by_xpath("//div[@class='col-4']//a"):
    
    urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
# Name of the Car
Names = []

# Variant of the Car
Variant = []

# Price of the car
Prices = []

# Fuel_type
Fuel = []

# KMS
Kilometers = []

# Engine
Purchase = []

# Owners 
Owners = []

# Transmission
Transmission = []

# SIts
Nonacc = []



for url in urls:
    driver.get(url)
    time.sleep(2)
    
    # Extracting Name from xpath
    try:
        name = driver.find_element_by_xpath("//h2[@class='_2geSF']")      
        Names.append(name.text)
    except NoSuchElementException:
        Names.append('-')
    
    
    
    # Extracting variant from Css Selector
    try:
        variant = driver.find_element_by_xpath("//span[@class='_18vo2']")      
        Variant.append(variant.text)
    except NoSuchElementException:
        Variant.append('-')
        
        
     
    # Extracting Price from Css Selector
    try:
        price = driver.find_element_by_xpath("//div[@class='text-right g-hPT']//strong")      
        Prices.append(price.text)
    except NoSuchElementException:
        Prices.append('-')
        
        
    
    # Extracting fuel from Css Selector
    try:
        fuel = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(4) > strong")      
        Fuel.append(fuel.text)
    except NoSuchElementException:
        Fuel.append('-')
        
   
    # Extracting Kilometers from Css Selector
    try:
        kms = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div > div > ul > li:nth-child(3) > strong")      
        Kilometers.append(kms.text)
    except NoSuchElementException:
        Kilometers.append('-')

        
     
    # Extracting engine from Css Selector
    try:
        purchase = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(9) > strong")      
        Purchase.append(purchase.text)
    except NoSuchElementException:
        Purchase.append('-')
        
        
    
    # Extracting owner count from Css Selector
    try:
        owner = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(2) > strong")      
        Owners.append(owner.text)
    except NoSuchElementException:
        Owners.append('-')
        
     
    # Extracting Transmission from Css Selector
    try:
        trans = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(6) > strong")      
        Transmission.append(trans.text)
    except NoSuchElementException:
        Transmission.append('-')
        
     
    # Extracting seats from Css Selector
    try:
        nas = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(1) > strong")      
        Nonacc.append(nas.text)
    except NoSuchElementException:
        Nonacc.append('-')
```

```python
cars_df = pd.DataFrame({'Name':Names,'Variant':Variant,'Fuel':Fuel,'Kilometers':Kilometers,'Purchase Date':Purchase,
                        'Owners':Owners,'Transmission':Transmission,'Accidential':Nonacc,
                        'Price':Prices})
```

```python
# Name of the Car
Names = []

# Variant of the Car
Variant = []

# Price of the car
Prices = []

# Fuel_type
Fuel = []

# KMS
Kilometers = []

# Engine
Purchase = []

# Owners 
Owners = []

# Transmission
Transmission = []

# SIts
Nonacc = []



for url in urls:
    driver.get(url)
    
    
    # Extracting Name from xpath
    try:
        name = driver.find_element_by_xpath("//h2[@class='_2geSF']")      
        Names.append(name.text)
    except NoSuchElementException:
        Names.append('-')
    
    
    
    # Extracting variant from Css Selector
    try:
        variant = driver.find_element_by_xpath("//span[@class='_18vo2']")      
        Variant.append(variant.text)
    except NoSuchElementException:
        Variant.append('-')
        
        
     
    # Extracting Price from Css Selector
    try:
        price = driver.find_element_by_xpath("//div[@class='text-right g-hPT']//strong")      
        Prices.append(price.text)
    except NoSuchElementException:
        Prices.append('-')
        
        
    
    # Extracting fuel from Css Selector
    try:
        fuel = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(4) > strong")      
        Fuel.append(fuel.text)
    except NoSuchElementException:
        Fuel.append('-')
        
   
    # Extracting Kilometers from Css Selector
    try:
        kms = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div > div > ul > li:nth-child(3) > strong")      
        Kilometers.append(kms.text)
    except NoSuchElementException:
        Kilometers.append('-')

        
     
    # Extracting engine from Css Selector
    try:
        purchase = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(9) > strong")      
        Purchase.append(purchase.text)
    except NoSuchElementException:
        Purchase.append('-')
        
        
    
    # Extracting owner count from Css Selector
    try:
        owner = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(2) > strong")      
        Owners.append(owner.text)
    except NoSuchElementException:
        Owners.append('-')
        
     
    # Extracting Transmission from Css Selector
    try:
        trans = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(6) > strong")      
        Transmission.append(trans.text)
    except NoSuchElementException:
        Transmission.append('-')
        
     
    # Extracting seats from Css Selector
    try:
        nas = driver.find_element_by_css_selector("body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div.col-8 > div > ul > li:nth-child(1) > strong")      
        Nonacc.append(nas.text)
    except NoSuchElementException:
        Nonacc.append('-')
        
        
```

```python
cars_df = pd.DataFrame({'Name':Names,'Variant':Variant,'Fuel':Fuel,'Kilometers':Kilometers,'Purchase Date':Purchase,
                        'Owners':Owners,'Transmission':Transmission,'Accidential':Nonacc,
                        'Price':Prices})
cars_df["Location"] = "Chennai"
df7=cars_df
```
