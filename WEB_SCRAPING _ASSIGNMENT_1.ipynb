```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Funtion to display all header tags
def h_tags(url):
    page=requests.get(url)
    print(page)
    soup= BeautifulSoup(page.content)
    h_tags=soup.find_all(['h1','h2','h3','h4','h5','h6'])
    print('All the header tags are as follows:','\n')
    for i in h_tags: 
        print('The header tag is:',i.name,'and heading is', i.text.replace('\n',''))
```

```python
#Function to display IMDB’s Top rated 100 movie’s data (i.e.Name,IMDB rating,Year of release) and save it in form of a CSV file.
def IMDB_top_100_movies(url):
    page=requests.get(url+'/chart/top/')
    #print(page)
    soup= BeautifulSoup(page.content)
    movie_name=soup.find_all('td',class_='titleColumn')
    movie_names=[]
    for i in movie_name:
        movie_names.append(i.get_text().replace('\n',''))
    
    movie=pd.DataFrame({})
    movie['raw']=movie_names
    movie['raw']=movie['raw'].str.replace('.',',').str.replace('(',',').str.replace(')','')
    movie['Rank'],movie['Name'],movie['Year of release']=movie['raw'].str.split(',',2).str
    rating=soup.find_all('td', class_='ratingColumn imdbRating')
    ratings=[]
    for i in rating:
        ratings.append(i.text.replace('\n',''))
    movie['Rating']=ratings
    movie.set_index(['Rank'])
    movie.drop('raw',axis=1,inplace=True)
    movie.iloc[0:100,].to_csv("IMDB's top 100 movies data.csv",index=False)
    print('IMDB top 100 movies are:','\n')
    return(movie.head(100))
```

```python
IMDB_top_100_movies('https://www.imdb.com')
```

```python
#Function to display IMDB’s Top rated 100 Indain movie’s data (i.e.Name,IMDB rating,Year of release) and save it in form of a CSV file.
def top_indian_movie_data(url):
    page=requests.get(url+'/india/top-rated-indian-movies')
    #print(page)
    soup= BeautifulSoup(page.content)
    movie_name=soup.find_all('td',class_='titleColumn')
    movie_names=[]
    for i in movie_name:
        movie_names.append(i.get_text().replace('\n',''))
    
    movie=pd.DataFrame({})
    movie['raw']=movie_names
    movie['raw']=movie['raw'].str.replace('.',',').str.replace('(',',').str.replace(')','')
    movie['raw']=movie['raw'].str.replace(' ','')
    movie['Rank'],movie['Name'],movie['Year of release']=movie['raw'].str.split(',',2).str
    rating=soup.find_all('td', class_='ratingColumn imdbRating')
    ratings=[]
    for i in rating:
        ratings.append(i.text.replace('\n',''))
    movie['Rating']=ratings
    movie.set_index(['Rank'])
    movie.drop('raw',axis=1,inplace=True)
    movie.iloc[0:100,].to_csv("IMDB's top 100 Indian movies data.csv",index=False)
    print("IMDB's top 100 Indain movies are:",'\n')
    return(movie.head(100))
```

```python
#Function to scraping book data of any 5 books from www.bookpage.com
def book_data(url):
    page=requests.get(url+"/reviews")
    #print(page)
    soup= BeautifulSoup(page.content)
    book_name=soup.find_all('h4',class_='italic')
    book_names=[]
    for i in book_name:
        book_names.append(i.get_text().replace('\n',''))
    author_name=soup.find_all('p',class_='sans bold')
    author_names=[]
    for i in author_name:
        author_names.append(i.text.replace('\n',''))
    genre_name=soup.find_all('p', class_='genre-links hidden-phone')
    genres=[]
    for i in genre_name:
        genres.append(i.text.replace('\n',''))
    review=soup.find_all('p', class_='excerpt')
    reviews=[]
    for i in review:
        reviews.append(i.text.replace('\n',''))
    book_data=pd.DataFrame({})
    book_data['Book Name']=book_names
    book_data['Author']=author_names
    book_data['Genre']=genres
    book_data['Review']=reviews
    book_data.replace('',np.nan,inplace=True)
    book_data['Book Name']=book_data['Book Name'].str.replace('\W',' ')
    book_data.dropna(axis=0,inplace=True)
    print('Book data of any 5 books are:')
    return book_data.head(5)
```

```python
book_data('https://bookpage.com')
```

```python
#Funtion to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in
def mobile_data(url):
    page=requests.get(url+'/s?i=electronics&bbn=1389401031&rh=n%3A976419031%2Cn%3A1389401031%2Cn%3A1389432031%2Cp_36%3A-2000000&dc&qid=1620551399&rnid=1389401031&ref=sr_nr_n_3')
    soup=BeautifulSoup(page.content)
    p_name=soup.find_all('span',class_='a-size-base-plus a-color-base a-text-normal')
    p_names=[]
    print(page)
    for i in p_name:
        p_names.append(i.text.replace('\n',''))
        
    price=soup.find_all('span',class_='a-price-whole')
    prices=[]
    for i in price:
        prices.append(i.text.replace('\n',''))
        
    rating=soup.find_all('span',class_='a-icon-alt')
    ratings=[]
    for i in rating:
        ratings.append(i.text.replace('\n',''))
        
    img_url=soup.find_all('img',class_='s-image')
    img_urls=[]
    for i in img_url:
        img_urls.append(i.get('src'))
        
    mob_data=pd.DataFrame({})
    mob_data['Product Name']=p_names
    mob_data['Price']=prices[0:len(p_names)]
    mob_data['Image URL']=img_urls[0:len(p_names)]
    mob_data['Average Rating']=ratings[0:len(p_names)]
    for i in range(0,len(mob_data['Product Name'])):
        mob_data['Product Name'][i]=mob_data['Product Name'][i].split('|')[0]
    print('Data of mobile phones under INR 20000 are as follows:')
    return mob_data
```

```python
#Funtion to display 7 days extended forecast National Weather Service website of USA
def weather_data(url):
    page=requests.get(url+'/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJOQcMDis2w')
    #print(page)
    soup= BeautifulSoup(page.content)
    period=soup.find_all('div',class_='col-sm-2 forecast-label')
    periods=[]
    for i in period:
        periods.append(i.text.replace('\n',''))
    desc=soup.find_all('div',class_='col-sm-10 forecast-text')
    description=[]
    for i in desc:
        description.append(i.text.replace('\n',''))
    weather=pd.DataFrame({'Period':periods,'Description':description})
    weather['Short Description'],weather['Temp in F']=weather['Description'].str.split(',',1).str
    weather['Temp in F'],weather['unn']=weather['Temp in F'].str.split('.',1).str
    weather['Temp in F']=weather['Temp in F'].str.extract('(\d+)')
    weather.drop(['unn'],axis=1,inplace=True)
    print('7 day extended forecast for San Francisco is as follows:')
    return weather
```

```python
weather_data('https://forecast.weather.gov')
```

```python
#Function to scrape cricket rankings from ‘www.icc-cricket.com’.
def Mens_cricket_data(url):
    page=requests.get(url+'/rankings/mens/team-rankings/odi')
    #print(page)
    soup= BeautifulSoup(page.content)
    data=soup.find('tr',class_='rankings-block__banner')
    f_team=data.text.replace('\n',':')
    f_team=f_team.split(':')
    f_team=[i for i in f_team if i]
    team_1=pd.DataFrame({'Rank':f_team[0],'Team':f_team[1],'Matches':f_team[3],'Points':f_team[4],'Rating':f_team[5]},index=['team1'])
    da=soup.find_all('tr',class_='table-body')
    d=[]
    for i in da:
        d.append(i.text.replace('\n',':'))
    t=[]
    for i in d:
        t.append(i.split(':'))
    teams=[]
    for i in t:
        teams.append([x for x in i if x])
    r=pd.DataFrame({})
    r['raw']=teams
    col=['Rank','Team','Matches','Points','Rating']
    for i in col:
        r[i]=pd.Series([1 for x in range(len(r.index))])
    for i in range(0,len(r)):
        r['Rank'][i]=r['raw'][i][0]
        r['Team'][i]=r['raw'][i][1]
        r['Matches'][i]=r['raw'][i][3]
        r['Points'][i]=r['raw'][i][4]
        r['Rating'][i]=r['raw'][i][5]
    r.drop(['raw'],axis=1,inplace=True)
    res=team_1.append(r)
    result=res.iloc[0:10].set_index(['Rank'])
    #res.set_index(['Rank'],inplace=True)
    page2=requests.get(url+'/rankings/mens/player-rankings/odi')
    soup= BeautifulSoup(page2.content)
    da=soup.find_all('td',class_='table-body__cell name')
    d=[]
    for i in da:
        d.append(i.text.replace('\n',''))
    team=soup.find_all('span',class_='table-body__logo-text')
    teams=[]
    for i in team:
        teams.append(i.text.replace('\n',''))
    rating=soup.find_all('td',class_='table-body__cell u-text-right rating')
    ratings=[]
    for i in rating:
        ratings.append(i.text.replace('\n',''))
    top=soup.find_all('div',class_='rankings-block__banner--name')
    top_p=[]
    for i in top:
        top_p.append(i.text.replace('\n',''))
    top=soup.find_all('div',class_='rankings-block__banner--nationality')
    top_p_team=[]
    for i in top:
        top_p_team.append(i.text.replace('\n',''))
    top_p_team
    items=[]
    for item in top_p_team:
        items.append(item.split())
    top_batsmen=pd.DataFrame({'Name':top_p})
    top_batsmen['Rating']=1
    top_batsmen['Team']=1
    for i in range(0,len(top_p_team)):
        top_batsmen['Rating'][i]=items[i][1]
        top_batsmen['Team'][i]=items[i][0]
    f_batsmen=pd.DataFrame(top_batsmen.iloc[0,]).transpose()
    r_players=pd.DataFrame({'Name':d,'Rating':ratings,'Team':teams})
    top_10_batsmen=f_batsmen.append(r_players.iloc[0:9,])
    top_10_batsmen['Rank']=range(1,11)
    top_10_batsmen.set_index(['Rank'],inplace=True)
    top_10_batsmen
    
    top_bolwer=pd.DataFrame(top_batsmen.iloc[1,]).transpose()
    top_10_bolwers=top_bolwer.append(r_players.iloc[9:18,])
    top_10_bolwers['Rank']=range(1,11)
    top_10_bolwers.set_index(['Rank'],inplace=True)
  
    print('Top 10 ODI teams in men’s cricket are:')
    print(result)
    print('\n\n\n Top 10 ODI Batsmen in men are: ')
    print(top_10_batsmen)
    print('\n\n\n Top 10 ODI Bowlers in men are: ')
    print(top_10_bolwers)
```

```python
Mens_cricket_data('https://www.icc-cricket.com')
```
