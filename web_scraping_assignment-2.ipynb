```python
#importing necessary libraries
import selenium
import pandas as pd
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
```

```python
#Program to scrape data for Data Analyst job position in banglore from naukri.com
def  job_bang():
    driver=webdriver.Chrome(r'chromedriver') #connecting to web driver
    driver.get("https://www.naukri.com/")
    #finding element by id for job searh bar and writing Data Analyst on it
    search_job= driver.find_element_by_id('qsb-keyword-sugg')
    search_job.send_keys('Data Analyst')

    #findind element for location search bar and entering location
    loc= driver.find_element_by_id('qsb-location-sugg')
    loc.send_keys('Bangalore')

    #finding search button and clicking on it
    search_btn=driver.find_element_by_xpath("//button[@class='btn']")
    search_btn.click()
    time.sleep(4)
    
    #extracting job titles for first 10 jobs
    titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
    job_titles=[] 
    for i in titles_tags:
        if len(job_titles)<10:
            job_titles.append(i.text)
            
    #extracting job location for first 10 jobs
    loc_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
    job_loc=[]
    for i in loc_tags:
        if len(job_loc)<10:
            job_loc.append(i.text)  
    
    #extracting company names of first 10 jobs
    company_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
    company_names=[]
    for i in company_tags:
        if len(company_names)<10:
            company_names.append(i.text)
     
    #extracting experience required
    exp_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")
    experience=[]
    for i in exp_tags:
        if len(experience)<10:
            experience.append(i.text)
            
    #Creating dataframe of the scraped data
    jobs=pd.DataFrame({'Job Title':job_titles,'Job Location':job_loc,'Company Name':company_names,'Experience Required':experience})
    
    return jobs
```

```python
def jobs_2():
    driver=webdriver.Chrome(r'chromedriver') #connecting to web driver
    driver.get("https://www.naukri.com/")
    
    #finding element by id for job searh bar and writing Data Scientist on it
    search_job= driver.find_element_by_id('qsb-keyword-sugg')
    search_job.send_keys('Data Scientist')
    
    #typing bangalore on location bar
    loc= driver.find_element_by_id('qsb-location-sugg')
    loc.clear()
    loc.send_keys('Bangalore')
    
    #clicking on search button
    search_btn=driver.find_element_by_xpath("//button[@class='btn']")
    search_btn.click()
    time.sleep(4)
    
    #getting urls of the job
    url=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
    urls=[]
    for i in url:
        urls.append(i.get_attribute('href'))
        
    #creating empty lists to store data
    job_titles=[]
    job_loc=[]
    company=[]
    full_desc=[]
    
    #Fetching the data from each url
    for i in urls[:10]:
        driver.get(i)
        time.sleep(2)
        try:
            job_titles.append(driver.find_element_by_xpath("//h1[@class='jd-header-title']").text)
        except NoSuchElementException:
            job_titles.append('NA')
        
        try:
            company.append(driver.find_element_by_xpath("//div[@class='jd-header-comp-name'][1]").text)
        except NoSuchElementException:
            company.append('NA')
    
        try:
            job_loc.append(driver.find_element_by_xpath("//div[@class='loc'][1]").text)
        except NoSuchElementException:
            job_loc.append('NA')
    
        try:
            full_desc.append(driver.find_element_by_xpath("//div[@class='dang-inner-html']").text)
        except NoSuchElementException:
            full_desc.append('NA')
            
    #Creating dataframe of the scraped data
    jobs=pd.DataFrame({'Job Title':job_titles,'Job Location':job_loc,'Company Name':company,'Full Job Description':full_desc})
    return jobs
```

```python
def glassdoor_4():
    #Connecting to web driver and going to the url
    driver= webdriver.Chrome(r"chromedriver")
    driver.get('https://www.glassdoor.co.in/index.htm')
    
    #Entering Data Scientist in job title field
    srch_bar=driver.find_element_by_id("sc.keyword")
    srch_bar.clear()
    srch_bar.send_keys('Data Scientist')
    
    #Entering Noida in location field
    loc_bar=driver.find_element_by_id("sc.location")
    loc_bar.clear()
    loc_bar.send_keys('Noida')
    
    #Clicking on search
    srch=driver.find_element_by_xpath("//button[@class='gd-ui-button ml-std col-auto SearchStyles__newSearchButton css-iixdfr']")
    srch.click()
    
    #Creating empty lists to store data
    company=[]
    days=[]
    rating=[]
    
    #Fetching company name, no of days ago job was posted and rating of the company
    for i in driver.find_elements_by_xpath("//a[@class=' css-l2wjgv e1n63ojh0 jobLink']"):
        company.append(i.text)
    for j in driver.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-mi55ob']"):
        days.append(j.text)
    for k in driver.find_elements_by_xpath("//div[@class='d-flex flex-column css-x75kgh e1rrn5ka3']"):
        rating.append(k.text)
        
    #Creating data frame and storing the fetched data
    jobs=pd.DataFrame({'Company Name':company[:10],'No. of days posted ago':days[:10],'Rating':rating[:10]})
    jobs=jobs.replace('','NA')
    return jobs
```

```python
def glassdoor_5():
    driver= webdriver.Chrome(r"chromedriver")#connecting to web driver
    driver.get('https://www.glassdoor.co.in/Salaries/index.htm')
    
    #Entering Data scientist in job title field
    srch_bar=driver.find_element_by_xpath("//input[@name='sc.keyword']")
    srch_bar.clear()
    srch_bar.send_keys('Data Scientist')
    
    #Entering Noida in location field
    loc_bar=driver.find_element_by_id("LocationSearch")
    loc_bar.clear()
    loc_bar.send_keys('Noida')
    
    #Clicking on search
    srch=driver.find_element_by_id('HeroSearchButton')
    srch.click()
    time.sleep(4)
    
    #Creating empty lists to store data
    company=[]
    no_of_sal=[]
    avg_sal=[]
    min_sal=[]
    max_sal=[]
    
    #Scraping the required data
    for i in driver.find_elements_by_xpath("//p[@class='m-0 ']"):
        company.append(i.text)
    for j in driver.find_elements_by_xpath("//p[@class='css-1uyte9r css-1kuy7z7 m-0 ']"):
        no_of_sal.append(j.text)
    for k in driver.find_elements_by_xpath("//div[@class='col-2 d-none d-md-flex flex-row justify-content-end']"):
        avg_sal.append(k.text.replace('\n',''))
    for l in driver.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values d-flex justify-content-between ']/span[1]"):
        min_sal.append(l.text)
    for a in driver.find_elements_by_xpath("//div[@class='common__RangeBarStyle__values d-flex justify-content-between ']/span[2]"):
        max_sal.append(a.text)
    
    #Creating and storing data in dataframe
    jobs=pd.DataFrame({'Comapany Name':company[:10],'No of Salaries':no_of_sal[:10],'Average Salary':avg_sal[:10],'Min Salary':min_sal[:10],'Max Salary':max_sal[:10]})
    return jobs
```

```python
def flipkart_6():
    driver= webdriver.Chrome(r"chromedriver")# connecting to the web driver
    driver.get('https://www.flipkart.com/')
    
    #removing the login pop up
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
    time.sleep(2)
    
    #Entering sunglasses in search field
    src=driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
    src.send_keys('sunglasses')
    
    #Clicking on search button
    driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
    time.sleep(4)
    
    #Creating empty lists to store data
    brand=[]
    desc=[]
    price=[]
    disc=[]
    
    #Fetching data from current page next two pages
    for i in range(0,3):
        for j in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
            if len(brand)<100:
                brand.append(j.text)
        for k in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
            if len(desc)<100:
                desc.append(k.text)
        for l in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
            if len(price)<100:
                price.append(l.text)
        for a in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']"):
            if len(disc)<100:
                disc.append(a.text)
        try:
            driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]").click()
            time.sleep(2)
        except:
            driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
            time.sleep(2)
        #Creating and storing data in data frame
    info=pd.DataFrame({'Brand':brand,'Product Description':desc,'Price':price,'Discount %':disc})
    return info
```

```python
def myntra_9():
    driver= webdriver.Chrome(r"chromedriver")#connecting to web driver
    driver.get('https://www.myntra.com/shoes')
    time.sleep(10)
    
    
    #select black in color checkbox
    c_filter=driver.find_elements_by_xpath("//label[@class='common-customCheckbox']")
    for i in c_filter:
        if ('Black') in i.text:
            i.click()
            break
    time.sleep(2)
    
    #selecting price range
    p_filter=driver.find_elements_by_xpath("//label[@class='common-customCheckbox vertical-filters-label']")
    for i in p_filter:
        if ('Rs. 6649 to Rs. 13099') in i.text:
            i.click()
            break
    
    brand=[]
    desc=[]
    price=[]
    
    #Scraping brand name, description and price of products
    for i in range(0,2):
        time.sleep(4)
        for i in driver.find_elements_by_xpath("//h3[@class='product-brand']"):
            brand.append(i.text)    
        for i in driver.find_elements_by_xpath("//h4[@class='product-product']"):
            desc.append(i.text)
        for i in driver.find_elements_by_xpath("//div[@class='product-price']/span[1]"):
            price.append(i.text)
        #Clicking on next button to go to next page
        driver.find_element_by_xpath("//a[@rel='next']").click()
     
    #Creating dataframe and storing the data
    data=pd.DataFrame({'Brand':brand,'Short Description':desc,'Price':price})
    a=data.Price.str.split('Rs.')
    #Selecting the current selling price
    for i in range(0,len(a)):
        data['Price'][i]=a[i][1]
    return data
```

```python
def amazone_10():
    driver= webdriver.Chrome(r"chromedriver")#connecting to web driver
    driver.get('https://www.amazon.in')
    time.sleep(4)
    
    #Enterting laptop in search bar and clicking the search button
    search_bar=driver.find_element_by_id('twotabsearchtextbox')
    search_bar.clear() #clearing the search bar
    search_bar.send_keys('Laptop')
    search_button=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']/span/input")
    search_button.click()
    time.sleep(4)
    
    #Selecting Intel Core i7 in CPU checkbox
    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span"):
        if i.text=='Intel Core i7':
            i.click()
            break
    time.sleep(4) 
    
    #Fetching url of the products
    url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    urls=[]
    for i in url:
        urls.append(i.get_attribute('href'))        
    
    title=[]
    rating=[]
    price=[]
    
    #Scraping the requird data for first 5 laptops
    for i in urls[:5]:
        driver.get(i)
        time.sleep(2)
        
        try:   
            for i in driver.find_elements_by_xpath("//span[@id='productTitle']"):
                title.append(i.text)
        except NoSuchElementException:
            title.append('NA')
            
        try: 
            for i in driver.find_elements_by_id("acrPopover"):
                a=i.get_attribute('title')
                t=[]
                t.append(a)
        
                #rating.append(i.get_attribute('title'))     
        except NoSuchElementException:
            rating.append('NA')
        rating.append(t[0])
        
        try:
            for i in driver.find_elements_by_xpath("//td[@class='a-span12']/span[1]"):
                price.append(i.text)
        except NoSuchElementException:
            rating.append('NA')
            
    #Enterting laptop in search bar and clicking the search button again 
    search_bar=driver.find_element_by_id('twotabsearchtextbox')
    search_bar.clear() #clearing the search bar
    search_bar.send_keys('Laptop')
    search_button=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']/span/input")
    search_button.click()
    time.sleep(4)
    
    #Selecting intel core i9 in CPU checkbox
    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span"):
        if i.text=='Intel Core i9':
            i.click()
            break
    time.sleep(4)   
    
    #Fetching url of the products
    url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    urls=[]
    for i in url:
        urls.append(i.get_attribute('href'))        
    
    #Scraping data for first 5 laptops with intel core i9
    for i in urls[:5]:
        driver.get(i)
        time.sleep(2)
        
        try:
            for i in driver.find_elements_by_xpath("//span[@id='productTitle']"):
                title.append(i.text)
        except NoSuchElementException:
            title.append('NA')
            
        try: 
            for i in driver.find_elements_by_id("acrPopover"):
                a=i.get_attribute('title')
                t=[]
                t.append(a)
             
        except NoSuchElementException:
            rating.append('NA')
        rating.append(t[0])
        try:
            for i in driver.find_elements_by_xpath("//td[@class='a-span12']/span[1]"):
                price.append(i.text)
        except NoSuchElementException:
            price.append('NA')
    
    #Creating dataframe and storing data 
    data=pd.DataFrame({'Title':title,'Rating':rating,'Price':price})
    return data
```
