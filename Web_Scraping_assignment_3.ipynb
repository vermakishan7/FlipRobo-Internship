Q1. Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‘guitar’. Then search for guitars.

```python

driver= webdriver.Chrome(r"chromedriver")
driver.get('https://www.amazon.in')
    
user_inp= input('Enter the product you want to search') #Taking input from the user that needs to be searched
search_bar=driver.find_element_by_id('twotabsearchtextbox')# locating the search bar
search_bar.clear() #clearing the search bar
search_bar.send_keys(user_inp+'s')# sending the user input to the search bar
search_button=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']/span/input") #Locating search button
search_button.click()#clicking on search button
```

Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on 
www.flipkart.com and scrape following details for all the search results displayed on 1st page. 
Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, 
“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display 
Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. 
Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe 
and CSV.

```python
driver= webdriver.Chrome(r"chromedriver")# connecting to the web driver
driver.get('https://www.flipkart.com/')

#removing the login pop up
driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
time.sleep(2)

#taking the user input
user_inp= input('Enter the product you want to search')
src=driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input") # locating search bar
src.send_keys(user_inp)# sending user input to search bar
#Clicking on search button
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
time.sleep(4)

#creating empty lists
Brand=[] 
smartphone_name=[]
Colour=[]
RAM=[] 
Storage=[]
Primary_Camera=[]
Secondary_Camera=[]
Display_Size=[]
Display_Resolution=[]
Processor=[]
Processor_Cores=[]
Battery_Capacity=[]
Price=[]
url=[]
#scraping ulrs of the prodcuts
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    url.append(i.get_attribute('href'))
#scraping prices of the products
for i in driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']"):
    Price.append(i.text)
for i in url:
    driver.get(i)
    time.sleep(4)
    for i in driver.find_elements_by_xpath("//span[@class='B_NuCI']"):# scraping brand name
        a=i.text.split(' ')
        Brand.append(a[0])
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()#clicking on read more
    #scraping details required using xpath
    a= driver.find_elements_by_xpath("//li[@class='_21lJbe']")
    b=driver.find_elements_by_xpath("//td[@class='_1hKmbr col col-3-12']")
    for i in range(0,len(b)):
        if b[i].text=='Color':
            Colour.append(a[i].text)
        if b[i].text=='Model Name':
            smartphone_name.append(a[i].text)
        if b[i].text=='Display Size':
            Display_Size.append(a[i].text)
        if b[i].text=='Resolution':
            Display_Resolution.append(a[i].text)
        if b[i].text=='Internal Storage':
            Storage.append(a[i].text)
        if b[i].text=='RAM':
            RAM.append(a[i].text)
        if b[i].text=='Primary Camera':
            Primary_Camera.append(a[i].text)
        if b[i].text=='Battery Capacity':
            Battery_Capacity.append(a[i].text)
#creating dataframe
df=pd.DataFrame({"Brand Name":Brand, "Smartphone name":smartphone_name, "Colour":Colour,"RAM":RAM, "Storage(ROM)":Storage,"Primary Camera":Primary_Camera, "Display Size":Display_Size, "Display Resolution":Display_Resolution, "Battery Capacity":Battery_Capacity, "Price":Price, "Product URL":url})
df
```

Q5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on 
google maps.

```python
driver= webdriver.Chrome(r"chromedriver")
driver.get('https://www.google.com/maps')
#getting input from user
user_inp=input('Enter city')
#sending the user input to search bar
driver.find_element_by_xpath("//input[@class='tactile-searchbox-input']").send_keys(user_inp)
driver.find_element_by_id('searchbox-searchbutton').click()
time.sleep(4)
#getting the current url
url=driver.current_url
#using regex sepearting the lat and long from the url
c= re.findall('@([0-9(.,)]+)',url)
for i in c:
    x=i.split(',')
print('Latitude: ',x[0],'\n Longitude: ',x[1])
```
